### Convert genre column from string to list 
#### ast.literal_eval make genre string type to list type

```python
original_movies = pd.read_csv("original_best_movies.csv")
original_movies["genre"] = original_movies["genre"].apply(ast.literal_eval)
original_movies.head(10)
```

### Get all genre types form original movies

```python
genres = set()
for list in original_movies["genre"]:
    for item in list:
        genres.add(item)

for item in genres:
    print(item)
```
